```python
df['senioridade'] = df['senioridade'].replace({
    'S√™nior': 'Senior',
    'Intermedi√°rio': 'Pleno',
    'Iniciante': 'Junior',

})

df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'M√©dia',
    'L': 'Grande'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'H√≠brido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```

## Exemplo de tratamento com dados nulos (dataframe teste)

```python
# utilizando a m√©dia de sal√°rios
import numpy as np
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Danielle', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# utilizando a mediana de sal√°rios
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
#Verificando a diferen√ßa entre utilizar m√©dia e mediana. A m√©dia acaba puxando o valor quando os dados tem outliers
```

```python
#Puxando os valores que estavam antes(segunda)
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [ 30, np.nan, np.nan, 28, 27]
})
df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
#Puxando os valores que estavam depois(Quinta)
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [ 30, np.nan, np.nan, 28, 27]
})
df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Danielle', 'Val'],
    'cidade':['S√£o Paulo', np.nan, 'Curitiba', np.nan, 'Bel√©m']
})
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('N√£o informado')
df_cidades
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title('Sal√°rio m√©dio por n√≠vel de senioridade')
plt.xlabel('N√≠vel de senioridade')
plt.ylabel('Sal√°rio m√©dio anual em d√≥lares')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df_limpo['usd'], bins = 50, kde=True)
plt.title('Distribui√ß√£o sal√°rios anuais')
plt.xlabel('Sal√°rio USD')
plt.ylabel('Frequ√™ncia')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['usd'])
plt.title("Boxplt sal√°rios")
plt.xlabel("Sal√°rio USD")
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']
plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue = 'senioridade')
plt.title("Boxplt Distribui√ß√£o por senioridade")
plt.xlabel("Sal√°rio USD")
plt.show()
```

```python
import plotly.express as px
```

```python
salario_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
salario_por_senioridade = salario_por_senioridade.sort_values('usd')

fig = px.bar(salario_por_senioridade, x='senioridade', y='usd',
             title='Sal√°rio m√©dio por n√≠vel de senioridade (Plotly)',
             labels={'senioridade': 'N√≠vel de senioridade', 'usd': 'Sal√°rio m√©dio anual em d√≥lares'})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(remoto_contagem, names='tipo_trabalho', values='quantidade',
title='Propor√ß√£o dos tipos de trabalho'
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(remoto_contagem, names='tipo_trabalho', values='quantidade',
title='Propor√ß√£o dos tipos de trabalho',
             hole = 0.5
)
fig.update_traces(textinfo='percent+label')
fig.show()
```
